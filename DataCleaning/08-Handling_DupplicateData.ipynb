# Handling Duplicate Data

```python
import pandas as pd
```

```python
data = {"name":["a","b","c","d","a","c"],"eng":[8,7,5,8,8,4],"hindi":[2,3,4,5,2,6]}
```

```python
df  = pd.DataFrame(data)
df
# when whole row is duplicate then data is called duplicate of another if there is only two columns are same but whole row is  not same so then they are not dupplicate.
```

```python
# duplicated()  : it is a function use find duplicates , true if duplicate exit false if no .

# df["Duplicates"] = df.duplicated()
# df
```

Remove duplicate Data 

```python
# Use drop_duplicates() : fxn for removing duplicates

# keep : parameter used to take which duplicate we want

# inplace parameter : not to create new change in the same dataset.

df.drop_duplicates(keep="first",inplace=True,)
df
```
