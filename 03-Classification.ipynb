# Classification

Machine learns from example and then try to classify new things into groups(labels,or categories)

Classification predicts labels unlike regression which predicts only numerical values.
Classification's output is a fix category 

# 1. Logistic Regression :


it is a classification model which is used to get or predict a binary outcome(0 or 1) using its input features .

```python
from sklearn.linear_model import LogisticRegression

X = [[0],[1],[2],[3],[4],[5]]
y = [0,0,1,1,1,1]

model = LogisticRegression()

model.fit(X,y)

hours = float(input("Enter hours of studying :\n"))

result = model.predict([[hours]])[0]

if result == 1:
    print(f"Based on hours {hours} , you are likely to Pass.")
else :
    print(f"Based on hours {hours} , you are likely to Fail.")
```

# 2. KNN  : K Nearest Neighbour

KNN predicts the output for a new data point by looking at the K closest points (neighbors) from the existing dataset.

majorty wins

```python
# Not so efficient for big data : slow 

from sklearn.neighbors import KNeighborsClassifier

X = [[180,7],
     [200,7.5],
     [250,8],
     [290,8.3],
     [300,8.5],
     [330,9],
     [360,9.5]]

# 0 = apple , 1 = Orange

y = [0,0,0,1,1,1,1]

model_k = KNeighborsClassifier(n_neighbors=3) # n_neighbors  means kitne neighbors ko check krega 

model_k.fit(X,y)

weight = float(input("Enter the weight in grams :"))
size = float(input("Enter the size in cm : "))

prediction = model_k.predict([[weight,size]])[0] # [0]  ka use : 

if(prediction == 0 ):
    print("this is likely to be an Apple ")
else :
    print("this is likely to an orange")
```
