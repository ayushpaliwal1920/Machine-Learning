# Apriori Algorithm 

Apriori Algorithm is a basic method used in data analysis to find groups of items that often appear together in large sets of data. It helps to discover useful patterns or rules about how items are related which is particularly valuable in market basket analysis.

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
pd.set_option("display.max_rows",500)
```

```python
dataset = pd.read_csv("groceries - groceries.csv")
dataset.head()
```

Handling missing values :

```python
print(dataset["Item 1"][0])
print(dataset["Item 5"][0])
```

```python
print(type(dataset["Item 1"][0]))
print(type(dataset["Item 5"][0]))
```

```python
dataset.columns
```

```python
print(dataset.shape)
```

```python
market = [] 

for i in range(0,dataset.shape[0]) :
    cust = []
    for j in dataset.columns :
        if type(dataset[j][i]) == str :
            cust.append(dataset[j][i])
    market.append(cust)
```

```python
new_market_list = []

for i in market:
    for j in i :
        new_market_list.append(j)
```

Most frequently selling byproduct : counting products occurences 

```python
import collections 
```

```python
products = collections.Counter(new_market_list)
```

```python
print("Keys :\n" , products.keys())
print("values :\n" , products.values())
```

```python
d = {"Item Name " : products.keys() , "values" : products.values()}
```

```python
df = pd.DataFrame(d)
```

```python
df.sort_values(by=["values"],ascending=False).head(10)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
new_df = pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

```python
new_df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(new_df, min_support = 0.05 , use_colnames = True , max_len = 3 ).sort_values(by = ["support"])
```
