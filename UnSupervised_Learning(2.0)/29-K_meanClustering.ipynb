# K Mean Clustering :

K-means Clustering: Algorithm, Applications, Evaluation ...K-means clustering is an unsupervised machine learning algorithm that partitions 
 observations into 
 distinct, non-overlapping clusters based on feature similarity. It iteratively minimizes within-cluster variance (inertia) by assigning data points to the nearest centroid and recalculating centroids until convergence. Key applications include customer segmentation, image compression, and anomaly detection

use k means when data is linearly seperable.

Uses Elbow Technique to find the number of clusters.

In elbow method we find the wcss : within cluster sum of squares 

Then make a graph btw number of clusters and wcss.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

dataset = pd.read_csv("Iris_raw.csv")

# dataset = dataset.drop(columns=["Id"])

# dataset.to_csv("Iris.csv",index=False)
dataset.head()
```

Check : Data is linearly seperable or not

```python
sns.pairplot(data=dataset)
plt.show()
```

Finding Number of Clusters : using wcss or elbow method

```python
from sklearn.cluster import KMeans
```

```python
# Finding the value of wcss : using 'intertia_' method

wcss = []

for i in range(2,21) :
    km = KMeans(n_clusters=i ,init='k-means++') # k-means++ uses for finding best clusters
    km.fit(dataset)
    wcss.append(km.inertia_) # finds the value of 'wcss' : .inertia_
```

```python
plt.figure(figsize=(10,5))
plt.plot([i for i in range(2,21)],wcss,marker = "o")
plt.xlabel("number of clusters")
plt.xticks([i for i in range(2,21)])
plt.ylabel("wcss")
plt.grid(axis="x")
plt.show()
```

Clustering : 


```python
kmn = KMeans(n_clusters=3)

dataset["Predict_Output"] = kmn.fit_predict(dataset)
```

```python
sns.pairplot(data=dataset ,hue="Predict_Output")
plt.legend(loc=5)
plt.show()
```
