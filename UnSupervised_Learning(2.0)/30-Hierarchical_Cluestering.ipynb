# Hierarchical Clustering :

Hierarchical clustering is an unsupervised machine learning algorithm that groups data points into a tree-like hierarchy of clusters, represented by a dendrogram.

 It requires no prior knowledge of cluster numbers, using distance metrics (e.g., Euclidean) and linkage methods (e.g., Single, Complete, Average) to merge or split data points

It is used when data is linearly seperable.

The Dendrogram is a tree like structure that is mainly used to store each step as  a memory that the HC algorithm performs .

The dendrogram plot the Y axis shows the Euclidean distances between the data points and the x axis shows all the data points of the given dataset.

HC has two approaches : Agglometrative and Divisive 

# Agglometrative :

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Iris_raw.csv")
dataset.head()
```

Check Linearly seperable or not : this works only on Linearly seperable data

```python
sns.pairplot(data=dataset)
plt.show()
```

Make Dendrogram : using scipy library

```python
import scipy.cluster.hierarchy as sc
```

```python
# parameters of dendrogram : z(Linkage)
# parameters of linkage  : dataset , method = "single","avegrage","centroid" or "complete" , metrics ="euclidean"(for claculating distance)

plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(dataset,method="single",metric="euclidean"))
plt.show()
```

We Found the nume=ber of cluster by observing the Dendrogram : read from copy or gfg...


number of clusters = 2.

Now Clustering 

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2 ,linkage="single")

dataset["Predict"] = ac.fit_predict(dataset)
dataset.head()
```

```python
sns.pairplot(data=dataset,hue="Predict")
plt.show()
```
