# Unsupervised Learning


A type of machine learning where algorithms analyze unlabeled data to discover hidden patterns, structures, or groupings without explicit human guidance or predefined output labels, contrasting with supervised learning's use of labeled data

it excels at tasks like clustering (segmenting customers), dimensionality reduction (simplifying data), and association rule mining (finding relationships), making it ideal for exploratory data analysis, customer segmentation, and anomaly detection in areas like finance and cybersecurity. 

# Clustering 

an unsupervised technique for grouping similar unlabeled data points into "clusters," where items within a cluster are more alike than items in other clusters, revealing inherent patterns without prior category knowledge, useful for customer segmentation, image analysis, and anomaly detection. It organizes data by similarity, using metrics like Euclidean distance to define groups, forming the basis for understanding data structure in fields from marketing to biology. 

It is not for prediction...It only groups similar type of data or finds patterns in data and groups it..

# K Mean Clustering 

K-Means clustering is an unsupervised machine learning algorithm that partitions unlabelled data into \(K\) distinct, non-overlapping clusters based on feature similarity. It iteratively minimizes the distance (usually Euclidean) between data points and their assigned cluster centroid, optimizing data organization for tasks like customer segmentation, image compression, and anomaly detection
