# Decision Tree Regression 

Decision tree regression is a non‑parametric supervised learning method used for predicting a continuous target variable. It models the relationship between features and a target by learning simple decision rules inferred from the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Salary_Data.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

Check Distribution :

```python
sns.pairplot(data=dataset)
plt.show()
```

Seperate data :

```python
x = dataset.iloc[:,[1,2]]
y = dataset["YearsExperience"]
```

train test :

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

Decision tree regressor :

in DTR there is crietien Mean square error , Rmse , Mean Absolute error not gini index or entropy

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

```python
dt.score(x_train,y_train)
```

```python
plot_tree(dt)
plt.show()
```
