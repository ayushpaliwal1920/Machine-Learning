# Backward and Forward Elimination (using MLXTEND)

Feature Selection in Machine Learning is the process of choosing the most relevant input features for a model and removing irrelevant or redundant ones.

It helps you build simpler, faster, and more accurate models—which is especially important as you move toward expert-level ML engineering.

choosing only important and necessary columns(features) for machine learning input is feature selection

# Foward Elimination

Forward Elimination is a wrapper-based feature selection technique where we start with zero features and add one feature at a time, choosing the feature that improves model performance the most.

# Backward Elimination

Backward Elimination is a wrapper-based feature selection technique where we start with all features and remove the least useful feature one by one, based on model performance or statistical significance.


```python
import pandas as pd
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LogisticRegression
from sklearn.impute import  SimpleImputer

dataset = pd.read_csv("loan.csv")
dataset.head(3)

# from sklearn.impute import SimpleImputer

# si = SimpleImputer(strategy="mean")
# arr = si.fit_transform(dataset_sk[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
#        'Credit_History']])

# new_dataset = pd.DataFrame(arr,columns=dataset_sk.select_dtypes(include="float64").columns)

# new_dataset.isnull().sum()
```

```python
# seperate data x and y

x = dataset.iloc[:,:-1]
y = dataset["Loan_Status"]

x.shape
```

```python
# one hot encoding 

x.drop(columns=["Loan_ID","LoanAmount","Loan_Amount_Term","Credit_History"],inplace=True)

x = pd.get_dummies(x,drop_first=True)


```

```python
x.isnull().sum()

x.shape
```

```python

lr = LogisticRegression()

# Sequential feature selector

fs = SequentialFeatureSelector(lr,n_features_to_select=10,direction="forward")
fs.fit(x,y)
```

```python
fs.feature_names_in_
```

```python
fs.tol
```
