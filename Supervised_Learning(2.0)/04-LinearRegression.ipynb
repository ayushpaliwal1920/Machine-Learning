# Linear Regression

Linear Regression is a supervised learning algorithm used to model the relationship between input features (X) and a continuous target variable (y) by fitting a best-fit straight line.

.. we use Simple Linear Regression when we have only one input feature to predict the output

ex - package by cgpa

..in this regression we plot a Regression Line on the graph of inputfeature vs outputfeature based on the given dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = {
    "cgpa" : [6.89,5.12,7.82,7.42,6.94,7.89,6.73,6.75,6.09,8.31,5.32,6.61],
    "package" :[3.26,1.98,3.25,3.67,3.57,2.99,2.6,2.48,2.31,3.51,1.86,2.6]
}

df = pd.DataFrame(data)
df.head()
```

```python
df.isnull().sum()
```

```python
# we always need data in two dimensions means rows and columns 
# here in one input feature case we need two [[]] for make it twoD.

x = df[["cgpa"]]
y = df["package"]
```

#Check the graph if linear regression is applicable or not by checking linearity .


```python

plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=df)
plt.show()

# if the direction of the graph is linear or in one direction then we can apply linear regression model
```

#train and test the model

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, random_state=42)

model = LinearRegression()

model.fit(x_train,y_train)

```

#regression line : y = mx +c 

```python
print("Slope of regression line :" ,model.coef_ )# this is the slope of the line

print("Intercept of regression line :" , model.intercept_) # this is the intercept of the line 
```

```python
# y = 0.51707425*x - 0.6980940518876317 # this is the equation of the line which we get from the slope and intercept
```

```python
model.predict([[5.12]])
```

#Now checking accuracy score of the model :


```python

model.score(x_test,y_test) # this is the r2 score of the model which is 0.8 means 80% accurate model

print("Model Score in percentage : ", model.score(x_test,y_test)*100)
```

#checking the regression line on the graph  :

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="package",data=df,label="org")
plt.plot(df["cgpa"] , model.predict(x),color="red" ,label = "predict line" )
plt.legend(["org" ,"predict line"])
plt.savefig("predict.jpg")
plt.show()
```
