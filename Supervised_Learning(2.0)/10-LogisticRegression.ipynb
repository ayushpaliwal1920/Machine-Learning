# Logistic regression : Binary classification (multiple input)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Placement.csv")
dataset.head()
```

check : is it linearly seperable ?  if yes then apply linear model if not then use another method

```python
sns.scatterplot(x="CGPA",y="IQ",data=dataset,hue="Placement") # # data is linearly seperable
plt.legend(loc = 1)
plt.show()
```

seperate data in dependent and independent data

```python
x = dataset.drop(columns = ["Placement" ,"Student_ID"])
y = dataset["Placement"]
```

train test splitting 

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()

lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[8.14,125]])
```

using mlxtend for making decision boundary : how logistic regression is seperaing its data

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```
