# SVM : Regression 

SVR is a regression technique that uses SVM for modeling and predicting continuous outcomes

In svr it searches the the most outer points for support vector and makes hyper plane using them and distance btw support vectors is d then d/2 is the distance where prediction line will be.

```python
import pandas as pd 
import numpy as no 
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("Placement.csv")
dataset.head()
```

```python
# Check Distribution :
plt.figure(figsize=(3,3))
sns.scatterplot(x="CGPA",y="IQ",data=dataset,hue="Placement")
plt.show()
```

```python
# Seperate data :

x = dataset.iloc[:,[1,2]]
y = dataset["Placement"]
```

```python
from sklearn.model_selection import train_test_split

x_train ,x_test,y_train,y_test  = train_test_split(x,y,random_state=20,test_size=0.2)
```

```python
from sklearn.svm import SVR

sv = SVR(kernel="linear") # use kernal using hit and trail method
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)
```

```python
sns.scatterplot(x="CGPA",y="IQ",data=dataset,hue="Placement")
plt.plot(dataset["CGPA"],sv.predict(x),color = "red")
plt.show()
```
