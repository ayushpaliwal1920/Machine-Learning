# Cross Validation :

Cross-validation is a robust resampling technique in machine learning used to evaluate model performance and prevent overfitting by training and testing a model on multiple, rotating subsets (folds) of the data.

a robust resampling technique in machine learning used to evaluate model performance and prevent overfitting by training and testing a model on multiple, rotating subsets (folds) of the data. It ensures every data point acts as both training and validation data, providing a more reliable estimate of how the model generalizes to unseen data than a single train-test split.

it validates the model efficiency by training it on the subset of input data and testing on previously unseen subset of input data.

# LeavePOut :

```python
from sklearn.model_selection import LeavePOut
```

```python
lp = LeavePOut(p=3)  # p ki value k hisab se split hoga
 
for train , test in lp.split(x_new,y_new) :
    print(train,test)
```

# KFold :
