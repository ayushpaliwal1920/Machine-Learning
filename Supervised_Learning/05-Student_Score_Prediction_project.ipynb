Note :

1. If X and y are numeric => Draw Scatter plot
2. If One is categorical data     => Draw BarPLot
3. If want to see Distribution shape => Draw histogram  or box plot


```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error
import numpy as np

data = pd.read_csv("student.csv")

X = data[["Hours"]]
y = data[["Score"]]

model = LinearRegression()
model.fit(X,y)

predicted_score = model.predict(X)

# evaluate

mae = mean_absolute_error(y,predicted_score)
mse = mean_squared_error(y, predicted_score)

rmse = np.sqrt(mse)

print("\nMAE :" , mae)
print("\nMsE :" , mse)
print("\nRMAE :" , rmse)

new_hour = float(input("enter hours : "))
new_pred = model.predict([[new_hour]])

print(f"\nPrediction for {new_hour} is {new_pred}")
```
