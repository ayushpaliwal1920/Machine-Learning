```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("student_performance.csv")
data
```

```python
# input model 

X = data[["weekly_self_study_hours"]]
y = data["total_score"]

# train model

model = LinearRegression()
model.fit(X,y)
predicted_score = model.predict(X)

# valid regression matrics

mae = mean_absolute_error(y,predicted_score)
mse = mean_squared_error(y,predicted_score)
rmse = np.sqrt(mse)

r2 = r2_score(y,predicted_score)

print("\nMae  :",round(mae,2)) # round = round off up to 2 figure
print("\nMse  :",round(mse,2))
print("\nRMae  :",round(rmse,2))
print("\nR^2 Score (Model Accuraacy)  :",round(r2,4)) # closer to 1 =  better


# histogram 

plt.figure(figsize=(10,6))
plt.hist(data["total_score"] , bins=30 , color='skyblue',edgecolor = 'black')
plt.grid()
plt.title("Distribution of Total Exam Score")
plt.xlabel("Final Exam Score")
plt.ylabel("Number of students")
plt.show()


# scatter plot  -> Regression line

plt.figure(figsize=(10,6))
plt.scatter(X , y ,color = 'blue' , label ="Actual Scores" )
plt.plot(X,predicted_score,color="red",label = "predicted_score(Regression Line)")
plt.grid()
plt.title("Model Prediction vs Actual Score")
plt.xlabel("Studied hours per week")
plt.ylabel("Final Output")
plt.show()

new_hours = 9
predicted_new_score = model.predict([[new_hours]])
print(f"predicted final score for {new_hours} hours is {predicted_new_score} score")
```
