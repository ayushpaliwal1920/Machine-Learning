# Regression : 

Regression is a machine learning/statistical method used to predict numbers.

```python
# Linear Regression – predicts using a straight-line relationship 
#  .finds  a pattern in old data , the make straight line form that data, then predict new or future data on the basis of the line.


#  fit() => trains from old data , It learns the relationship between input (X) and output (y).
# predict() => means using the trained model to make predictions

# Notes :
#  1. its not about the line its about the story .
#  2. you can start from 5 rows
#  3. accuracy is not always the goal.

import pandas as pd

from sklearn.linear_model import LinearRegression

# old input data :

X = [[1],[2],[3],[4],[5]]

# old output data :

y = [[20],[30],[50],[75],[86]]


model = LinearRegression()

model.fit(X,y) # x is old input and y is old output data.

hours = 2.5

predicted_marks = model.predict([[hours]])  # we passesd a 2d list , data should be in form of 2d

print(f"Based on your hours {hours} you may have scored around {predicted_marks} marks")
```
