# Data Preprocessing : Handling missing data


means handling missing data and data cleaning.

Notes : 1.Don't drop data until it is not fillable or try to fill data first.

NOtes : 2.fill numeric data with mean and median vlaues.

Notes : 3.Check how many precent data is missing before cleaning.

```python
import pandas as pd

dataf = {
    'Name':['Pavan','Lalit','Ayush','Rohan','Ohm'],
    'Age' :[25,None,56,54,None],
    'Salary':[25000,52000,53333,None,10000]
}

df = pd.DataFrame(dataf)
print("\n\nOriginal Data frame :")
print(df)

# Cheking null values :

print(df.isnull().sum())


# missing data percentage :

print("\n\nData missing percentage")
print(df.isnull().mean()*100)

# dropping column and rows whose has null values :

df_drop = df.dropna()
print(df_drop)

# Filling null values with mean values : and Inplace means modify the original DataFrame, NOT create a new one.

df['Age'].fillna(df["Age"].mean(),inplace=True)
df['Salary'].fillna(df["Salary"].mean(),inplace=True)
df
```

# Encoding Categorical Data : label Encoding

```python
# :Label encoding : it is a data preprocessing technique used in machine  learning to convert categorical text labels into numeric valuess.

# fittransform() - → Learn from the data and  Convert categories → numbers

from sklearn.preprocessing import LabelEncoder
import pandas as pd

df = pd.read_csv("large_sample_data.csv")

df_label = df.copy()

# creating a object of LabelEncoder

le = LabelEncoder()

# convert categorical data into numerical data
df_label['Gender_Encoded'] = le.fit_transform(df_label["Gender"]) 
df_label["Passed_Encoded"] = le.fit_transform(df_label["Passed"])

# Now showing encoded data
print("\nLabel Encoded Data :\n")
print(df_label[["Name","Gender","Gender_Encoded","Passed","Passed_Encoded"]].head(10))
```

# Encoding Categorical Data : One Hot Encoding

```python
# One hot Encoding : Convert categorical value into multiple columns and each will be representing one possible value.It converts each category into a new column with values 0 or 1.

# It is useful because when we have multiple unlabled data ex Cities , cars etc. 


# get_dummies() :It converts categorical columns into multiple columns containing 0s and 1s.used to perform One-Hot Encoding quickly and easily.

df_encoded = pd.get_dummies(df,columns=["City"],dtype=int).head(20)
print("\nOne-hot Encoded Data(City ):\n")
print(df_encoded) # it is in form of True and false

# Converting true & false into binary data :  there are two ways :: 
# 1. Just add dtype=int in get_dummies() function.
# 2. df_encoded = df_encoded.astype(int)

```

# Feature Scaling 

It ensures that all features (columns) have similar ranges, so the model learns correctly.

Two Main Types of Feature Scaling :
1. Standardization(StandardScaler) =>Mean becomes 0 ,Standard deviation becomes 1 ,Useful when data has outliers.

2. Normalization (Min-Max Scaling): Converts data → 0 to 1 range,Sensitive to outlier.Used in Deep Learning & Neural Networks

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler

data ={
    'StudyHours': [1,2,3,4,5],
    'TestScore' : [40,50,60,70,80]
 }


# StandardScaler :

df_ex = pd.DataFrame(data)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_ex)

# MinMaxScaler :

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df_ex)
```

Splitting : dividing dataset into different parts so that your machine model can learn.validate and test properly.




we split data Because:

The model must learn from one part → Training set

The model must be tested on unseen data → Test set

This helps avoid overfitting and gives true accuracy.

Most Common Split: Train–Test Split

Usually:

80% → Training set

20% → Test set

```python
import pandas as pd
# from sklearn import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split

data ={
    'StudyHours': [1,2,3,4,5],
    'TestScore' : [40,50,60,70,80]
 }

df = pd.DataFrame(data)

# Standard Scaler  :  formula => z = (x - mean)/std_deviation(of a colm)

standard_scaler = StandardScaler()
standard_scaled = standard_scaler.fit_transform(df) # fit and trandform

print("Standard Saler Output :\n")
print(pd.DataFrame(standard_scaled,columns=['StudyHours','TestScore']))

# Min_Max Scaler : formula => scaled = (x-min)/(max-min)

minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

print("\nMin Max Scaled Output :\n")
print(pd.DataFrame(minmax_scaled,columns=["StudyHours","testScore"]))

# train_test_split :

x = df[["StudyHours"]] # input
y = df[["TestScore"]]  # output

x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.2,random_state=42) # test_size=0.2 Means 20% of data goes to testing And 80% goes to training

# random_state=42 => Ensures the split is same every time,If you run the code again, the data will split identically


print("\nTraining Data : \n")
print(x_train)

print("\nTesting Data : \n")
print(x_test)

print("\nTraining Data : \n")
print(y_train)

print("\nTesting Data : \n")
print(y_test)
```
